## Mini-teste - Readmissões em Hospitais Medicare

Em outubro de 2012, o Centro de Serviços Medicare e Medicaid (CMS) do governo dos EUA começou a reduzir os pagamentos do Medicare para hospitais do Sistema de Pagamento Prospectivo de Pacientes Internados com readmissões em excesso. 

O excesso de readmissões é medido por uma razão, dividindo-se o número de readmissões “previstas” de 30 dias de um hospital para ataque cardíaco, insuficiência cardíaca e pneumonia, pelo número que seria “esperado”, com base em um hospital com características similares. Uma proporção maior que 1 indica readmissões em excesso.

* [Dicionário de Dados](https://data.cms.gov/provider-data/sites/default/files/data_dictionaries/hospital/HospitalCompare-DataDictionary.pdf)
* [Pandas Chart Visualization](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html)

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/amkaris/EDA/master/cms_hospital_readmissions.csv')
```

### 1. Crie uma coluna `True Admission Rate`, dividindo a coluna `Number of Readmissions` pela `Number of Discharges`

```python
df['Number of Discharges'][df['Number of Discharges'] == 'Not Available'] = np.nan
```

```python
df['True Admission Rate'] = (df['Number of Readmissions'].astype(float) / df['Number of Discharges'].astype(float)) * 100
```

```python
df.head()
```

### 2. Plote um boxplot da diferença entre `True Admission Rate` e `Expected Readmission Rate`. Disserte.


```python
sns.boxplot(data=df[["True Admission Rate", "Expected Readmission Rate"]], orient="v")
```

Podemos verificar que ambas métricas possuem medianas parecidas, porém a métrica "True Admission Rate" tem mais variância e mais outliers com valores mais altos. Podemos concluir que apesar da mediana ser parecida, "True Admission Rate" é mais alto que "Expected Readmission Rate".

### 3. Plote um scatterplot das colunas `Predicted Readmission Rate` e `Expected Readmission Rate`, com as cores dos pontos definidas pela coluna `Measure Name`. Disserte.

```python
set(df["Measure Name"])
```

```python
sns.scatterplot(data=df, x="Predicted Readmission Rate", y="Expected Readmission Rate", hue="Measure Name")
```

(Measure dictionary: https://www.ahd.com/definitions/qual_collection_periods.php)

- A doença "READM-30-HIP-KNEE-HRRP" (Hip/Knee Surgery) possui os menores índices, tanto para "Predicted" quanto para "Expected", ficando entre 0% e 10%
- A doença "READM-30-HF-HRRP" (Heart Failure) possui os maiores índices, dada a maior gravidade quando comparada às outras medidas. Também tem mais variância que as outras.
- A doença "READM-30-COPD-HRRP" (Chronic obstructive pulmonary disease) está apenas abaixo de "Heart Failure" em nível de gravidade.
- A doença "READM-30-PN-HRRP" (Pneumonia) é a segunda menos grave.
- A doença "READM-30-AMI-HRRP" (Heart Attack) possui alta variância.

### 4. Plote um scatterplot de `Excess Readmission Ratio` por `Number of Discharges`. Disserte.

```python
df["Number of Discharges"] = df["Number of Discharges"].astype(float)
```

```python
df['Excess Category'] = (df['Excess Readmission Ratio'] > 1).map({False: "Not excess", True: "Excess"})
```

```python
sns.scatterplot(data=df, x="Excess Readmission Ratio", y="Number of Discharges", hue='Excess Category')
```

Podemos ver que os casos com excesso de readmissão (> 1) costumam ter baixo número de dispensas.

### 5. Encontre os 5 estados com maior `Excess Readmission Ratio` e os 5 estados com menor `Excess Readmission Ratio`. Filtre o dataframe com estes 10 estados. Plote um gráfico com um boxplot por estado, considerando a coluna `Excess Readmission Ratio`

```python
states_excess = pd.DataFrame(df.groupby(['State'])['Excess Readmission Ratio'].mean().reset_index())
```

```python
top_5 = list(states_excess.sort_values('Excess Readmission Ratio')['State'].head(5))
```

```python
bottom_5 = list(states_excess.sort_values('Excess Readmission Ratio')['State'].tail(5))
```

```python
states_10 = []
states_10.extend(top_5)
states_10.extend(bottom_5)
```

```python
mask = df['State'].isin(states_10)
df_plot = df[mask]
```

```python
sns.boxplot(data=df_plot, x="State", y='Excess Readmission Ratio', orient="v")
```
